```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
t1 = np.arange(-50.0, 50.0, 0.1)
def f(t):
    return 0.2 * np.power(t,4) - 407 * np.power(t, 2) - 1129 * t + 134
        #0.5*np.exp(-t) + np.cos(np.power(t, 0.5)) + 0.05 * np.exp(np.power(np.abs(t), 0.5)) +  + 14



plt.figure()
plt.subplot(212)
plt.plot(t1, f(t1))

f(-9.0)
```

```python
t2 = np.arange(-10.0, 10.0, 0.1)
def f(t):
    return np.log(1+np.exp(t+2))
#-np.log(1+np.exp(t+1))+np.log(1+np.exp(t+2))
#-\
#           np.log(1+np.exp(t+3))
        #np.cos(np.power(np.abs(t), 0.5))
        #0.5*np.exp(-t) + np.cos(np.power(t, 0.5)) + 0.05 * np.exp(np.power(np.abs(t), 0.5)) +  + 14



plt.figure()
plt.subplot(212)
plt.plot(t2, f(t2))
plt.show()
f(-9.0)
```

```python
import random as rn
```

```python
dataset = []
for i in range (-50,50):
    for j in range (10):
        x = rn.uniform(i, i+1)
        y = f(x)
        dataset.append([x,y])
print(dataset)
datasetPermut = np.random.permutation(dataset)
print(datasetPermut)
datasetPermutNoise = datasetPermut.copy()
for k in range(len(datasetPermutNoise) - 1):
    datasetPermutNoise[k,0] = datasetPermutNoise[k,0] + rn.uniform(0.01, 0.1)
print(datasetPermutNoise)
import csv
with open('dataset_1000.csv','w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(datasetPermut)

with open('dataset_noise_1000.csv','w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(datasetPermutNoise)
```
