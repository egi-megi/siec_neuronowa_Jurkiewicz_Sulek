```python
import numpy as np
import matplotlib.pyplot as plt
import json

```

```python
with open('out/losses_2layers_100neurons_100000dataset_elu.json') as json_file:
    data_elu = json.load(json_file)
with open('out/losses_2layers_100neurons_100000dataset_selu.json') as json_file:
    data_selu = json.load(json_file)
with open('out/losses_2layers_100neurons_100000dataset_swish.json') as json_file:
    data_swish = json.load(json_file)
with open('out/losses_2layers_100neurons_100000dataset_sigmoid.json') as json_file:
    data_sigmoid = json.load(json_file)
with open('out/losses_2layers_100neurons_100000dataset_tanh.json') as json_file:
    data_tanh = json.load(json_file)
```

```python
elu=data_elu['val_loss']
selu=data_selu['val_loss']
swish=data_swish['val_loss']
sigmoid=data_sigmoid['val_loss']
tanh=data_tanh['val_loss']
x=np.arange(1,len(elu)+1)
```

```python
fig, ax = plt.subplots()
plt.xlabel("Liczba epok")
plt.ylabel("Błąd średniokwadratowy")
plt.ylim((0.0,0.02))
ax.plot(x, elu, 'k', label='elu',color = '#00FF00')
ax.plot(x, selu, 'k', label='selu',color = '#FF00FF')
ax.plot(x, swish, 'k', label='swish', color = '#0000FF')
ax.plot(x, sigmoid, 'k', label='sigmoid',color = '#FF0000')
ax.plot(x, tanh, 'k', label='tanh',color = '#000000')
legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.savefig('plots/MSE_epoch_plot.jpg',bbox_inches='tight',pad_inches=0.1)
```
